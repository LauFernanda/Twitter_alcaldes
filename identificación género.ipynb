```python
import pandas as pd
from gender_detector import gender_detector as gd
import unicodedata
from collections import Counter
from googlesearch import search
```

```python
df=pd.read_excel("./nombres_alcaldes.xlsx")
```

```python
df.columns
```

```python
detector = gd.GenderDetector('us')
```

```python
nombres=df["NOMBRES  Y  APELLIDOS"].tolist()
nombres_sep=[]
for nombre in nombres:
    x=unicodedata.normalize('NFKD', nombre).encode('ASCII', 'ignore').decode("utf-8") 
    nombres_sep.append(x.split(" "))
```

```python
 nombres_sep
```

```python
Genero={}
for nombre in nombres:
    Genero[nombre]={}
    x=unicodedata.normalize('NFKD', nombre).encode('ASCII', 'ignore').decode("utf-8") 
    nombre_sep=x.split(" ")
    try:
        gender_id=detector.guess(nombre_sep[0])
        if gender_id=='unknown':
            gender_id=detector.guess(nombre_sep[1])
        
        Genero[nombre]["Género Identificado"]=gender_id
    except:
        print(nombre)
```

```python
Counter([Genero[k]["Género Identificado"] for k in Genero.keys()])
```

```python
unknown_indexes=[i for i, j in enumerate(Genero) if j == 'unknown']
```

```python
unknown_gender=[nombres_sep[i] for i in unknown_indexes]
```

```python
unknown_gender
```

```python
for nombre in unknown_gender:
    try:
        Genero.append(detector.guess(nombre[0]))
    except:
        print(nombre)
```

```python
##Búsqueda usuario Twitter###
query=nombres[0]+" twitter"
print(query)
response = search(query, tld="co.in", num=1, stop=1, pause=2)
for result in response:
    print(result)
```

```python
usuario_twitter=[]
for nombre in nombres:
    query=nombre+" twitter"
    response = search(query, tld="co.in", num=1, stop=1, pause=2)
    for result in response:
        usuario_twitter.append(result)
    
```
